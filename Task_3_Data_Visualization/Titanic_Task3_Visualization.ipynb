### Objective
The objective of this project is to visualize the Titanic dataset using different charts to identify patterns, trends, and relationships between variables.

**Tools Used:**

- Pandas
- Matplotlib
- Seaborn

**Dataset:** Titanic Dataset (train.csv)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("train.csv")
```

```python
df.head()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Survived', data=df)
plt.title("Survival Count")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Sex', data=df)
plt.title("Gender Distribution")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Pclass', data=df)
plt.title("Passenger Class Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Age'], bins=20, kde=True)
plt.title("Age Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Fare'], bins=20, kde=True)
plt.title("Fare Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Pclass', y='Age', data=df)
plt.title("Age vs Passenger Class")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.barplot(x='Sex', y='Survived', data=df)
plt.title("Survival Rate by Gender")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
df['Pclass'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%',
    figsize=(6,6)
)
plt.title("Passenger Class Percentage")
plt.ylabel("")
plt.show()
```

# Conclusion

- Most passengers belonged to Passenger Class 3.
- Female passengers had a higher survival rate than male passengers.
- Most passengers were between 20 and 40 years old.
- Fare values were positively skewed with a few high-value outliers.
- Heatmap shows relationships among numerical variables.
